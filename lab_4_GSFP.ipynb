```python
# оценка стоимости опциона колл на отношение цен двух активов
def quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call):
    theta = 1
    if is_call == False:
        theta = -1
    sigma_hat = np.sqrt(sigma1 ** 2 + sigma2 ** 2 - 2 * rho * sigma1 * sigma2)

    d1 = (np.log(S1 / S2 / K) + (b1 - b2 - 0.5 * (sigma1 ** 2 - sigma2 ** 2)) * T) / \
         (sigma_hat * np.sqrt(T))
    d2 = d1 + sigma_hat * np.sqrt(T)
    F = S1 / S2 * np.exp((b1 - b2 +  sigma2 * (sigma2 - rho * sigma1)) * T)
    option = theta * np.exp(-r * T) * (F * norm.cdf(theta * d2) - K * norm.cdf(theta * d1))

    return option
```

```python
import numpy as np
import pandas as pd
from scipy.stats import norm

sigma1 = 0.3
sigma2 = 0.4
b1 = 0.05
b2 = 0.03
r = 0.07
S1 = 130
S2 = 100
T_range = [0.25, 0.5]
rho_range = [-0.5, 0, 0.5]
K_range = list(np.arange(0.1, 3.1, 0.1))

def quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call):
    theta = 1 if is_call else -1
    sigma_hat = np.sqrt(sigma1**2 + sigma2**2 - 2 * rho * sigma1 * sigma2)
    d1 = (np.log(S1 / (S2 * K)) + (b1 - b2 - 0.5 * (sigma1**2 - sigma2**2)) * T) / (sigma_hat * np.sqrt(T))
    d2 = d1 + sigma_hat * np.sqrt(T)
    F = (S1 / S2) * np.exp((b1 - b2 + sigma2 * (sigma2 - rho * sigma1)) * T)
    option = theta * np.exp(-r * T) * (F * norm.cdf(theta * d2) - K * norm.cdf(theta * d1))
    return option

def calculate_forward_price(S1, S2, b1, b2, T):
    F = (S1 / S2) * np.exp((b1 - b2) * T)
    return F

parity_df = pd.DataFrame([
    {
        'K': K,
        'rho': rho,
        'T': T,
        'Call_Price': quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call=True),
        'Put_Price': quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call=False),
        'Forward_Price': calculate_forward_price(S1, S2, b1, b2, T),
    }
    for K in K_range
    for rho in rho_range
    for T in T_range
])

parity_df['C_minus_P'] = parity_df['Call_Price'] - parity_df['Put_Price']
parity_df['Discounted_F_minus_K'] = np.exp(-r * parity_df['T']) * (parity_df['Forward_Price'] - parity_df['K'])

parity_df['Parity_Difference'] = parity_df['C_minus_P'] - parity_df['Discounted_F_minus_K']

max_difference = parity_df['Parity_Difference'].abs().max()
mean_difference = parity_df['Parity_Difference'].abs().mean()

print(f"Максимальное абсолютное отклонение: {max_difference:.10e}")
print(f"Среднее абсолютное отклонение: {mean_difference:.10e}")
```

```python
S = [100.0, 100.0]
K = 1.0
sigma = [0.18, 0.15]
q = [0.04, 0.03]
rho = 0.75
r = 0.05
T = 1.0

def quotient_price_Zhang(option_type, S, K, rho, sigma, r, q, T):
    b = [r - q_i for q_i in q]
    
    sigma_hat = np.sqrt(sigma[0]**2 + sigma[1]**2 - 2 * rho * sigma[0] * sigma[1])
    numerator = np.log(S[0] / (S[1] * K)) + (b[0] - b[1] - 0.5 * (sigma[0]**2 - sigma[1]**2)) * T
    denominator = sigma_hat * np.sqrt(T)
    d1 = numerator / denominator
    d2 = d1 + sigma_hat * np.sqrt(T)

    F = (S[0] / S[1]) * np.exp((b[0] - b[1] + sigma[1] * (sigma[1] - rho * sigma[0])) * T)
    
    if option_type == 'call':
        price = np.exp(-r * T) * (F * norm.cdf(d2) - K * norm.cdf(d1))
    elif option_type == 'put':
        price = np.exp(-r * T) * (K * norm.cdf(-d1) - F * norm.cdf(-d2))
    else:
        raise ValueError("Неверный тип опциона. Используйте 'call' или 'put'.")
    
    return price
```

# Выводы

В ходе анализа ценовых зависимостей опционов колл на максимум двух рискованных активов от коэффициента корреляции (\(\rho\)) были получены следующие результаты:

## Рост цены опциона с увеличением \(\rho\)

При повышении коэффициента корреляции \(\rho\) два базовых актива становятся более склонными к совместному движению. Это означает, что если один из активов демонстрирует рост, вероятность того, что второй актив также будет расти, увеличивается. В контексте опциона колл на максимум, это приводит к следующим эффектам:

- **Увеличение вероятности высоких значений максимума:** Поскольку активы движутся синхронно, если один актив достигает высокой цены, второй, вероятно, будет следовать за ним, увеличивая значение \(\max(S_1, S_2)\).
  
- **Снижение диверсификационных выгод:** При низкой корреляции активы могут двигаться независимо друг от друга, что иногда приводит к ситуациям, когда один актив растет, а другой падает. Это может ограничивать рост максимума, поскольку не оба актива одновременно достигают высоких значений.

В результате, при высоком \(\rho\) ожидаемая стоимость опциона колл на максимум возрастает, так как вероятность получения более высоких выплат увеличивается.

- **Высокая корреляция (\(\rho \approx 1\)):** Активы движутся в унисон. Если один актив показывает хороший рост, второй с высокой вероятностью тоже вырастет. Это делает максимум двух активов более предсказуемым и, как следствие, увеличивает ценность опциона колл на максимум.
  
- **Низкая или отрицательная корреляция (\(\rho \leq 0\)):** Активы движутся независимо или в противоположных направлениях. Это увеличивает неопределенность относительно того, какой из активов окажется выше, что может ограничивать максимальное значение и, соответственно, ценность опциона.

На построенном графике зависимости цены опциона колл от коэффициента корреляции видно, что с увеличением \(\rho\) цена опциона возрастает. Это подтверждает теоретическое ожидание, что более высокая корреляция между активами приводит к увеличению стоимости опциона на максимум.

Коэффициент корреляции \(\rho\) играет ключевую роль в оценке стоимости опционов колл на максимум двух рискованных активов. Повышение \(\rho\) увеличивает ценность опциона за счет повышения вероятности одновременного роста обоих активов, что приводит к более высокому значению максимума и, следовательно, к большей выплате по опциону. Это подчеркивает важность учета корреляционных зависимостей при оценке и управлении опционами на комбинации активов.
